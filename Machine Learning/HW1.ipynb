```python
data['gender'].unique()
```

```python
#Столбец gender переведем в значение 0 - женский  и 1 - мужской
data['gender'] = data['gender'].replace('Male', 1).replace('Female', 0)
data['gender'].unique()
```

```python
#Сделаем то же самое со столбцом income, >50K заменим на 1, <=50K тыс заменим на 0
data['income'] = data['income'].replace('>50K', 1).replace('<=50K', 0)
data['income'].value_counts() # Проверка на сохранность строк в датасете
```

```python
# выберем признаки
selectedColumns = data[ [ 'gender', 'education', 'income' ] ]

# столбец education является категориальной переменной
# переведем ее в значения 0 и 1, добавив столбцы с соответствующими названиями
X = pd.get_dummies( selectedColumns, columns = [ 'education' ] )

# столбец income является целевой переменной, удаляем его из X
del X['income']
X.head()
```

```python
y = pd.Series( data['income'] )
y.head()
```

```python
model = LogisticRegression()
```

```python
# обучаем модель

model.fit( X, y )
predictions = model.predict_proba( X )
```

```python
predictions[:5]
```

```python
for pred, y_true in zip( predictions[:, 1], y ):
    print(pred, y_true)
```

```python
model.score(X, y)
```
