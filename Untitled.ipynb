Даны вектора x=(1,1), y=(2, 0) и z = (0,2)

Надо:
Изобразить вектора на экране

Найти угол между векторами x и y

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
```

```python
#Изобразить вектора на экране
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
ax.scatter(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
#Изобразить точку x + y + z
c = x+y+z
plt.scatter(c[0],c[1])
```

```python
#Найти угол между векторами x и y
def angleBetween(a, b, deg = True):
    print('Первый вектор:', a)
    print('Второй вектор:', b)
    c = a.dot(b)
    result = np.arccos(a.dot(b)/(np.sqrt(a[0]**2+a[1]**2)*np.sqrt(b[0]**2+b[1]**2)))
    print('Радианы:', result)
    if deg:
        result = int(np.degrees([result.real])[0]) # переводим в градусы
    return result
```

```python
angleBetween(x,z)
```
